```python
# created funct

def user_vector(ratings,movie_vector):
    vector =[]
    for user in range(1,944):
        u1 = ratings[ratings['user_id']==user][['movie_id','rating']]
        u1_list=[]
        for i in u1['movie_id']:
            f = movie_vector[i].values
            u1_list.append(f)
        uu1 = pd.DataFrame(u1_list,columns=range(1,19))
        vec=[]
        for j in range(1,19):
            g = sum(uu1[j])
            vec.append(g/len(u1))
        vector.append(vec)
        
    return vector
```

```python
vector = user_vector(ratings,movie_vector)
vector
```

```python
user_vector = pd.DataFrame(vector)
user_vector
```

```python
# dot mat
content_vector = np.dot(user_vector,movie_vector)
content_vector # to get content vector we need uservec * movie vec
```

```python
content_vector = pd.DataFrame(content_vector)
content_vectors = content_vector.T
content_vectors
```

```python
content_vectors.to_csv("content_vector.csv",index=False)
```

```python
def content_based(ratings,content_vectors,userid,thresh):
    u2=ratings[ratings['user_id']==userid]['movie_id'].values
    val=content_vectors[content_vectors[userid]>(content_vectors[userid].max()-thresh)]
    recom=[]
    for i in val.index:
        if(i in u2):
            pass
        else:
            recom.append(i)
    return recom
```
